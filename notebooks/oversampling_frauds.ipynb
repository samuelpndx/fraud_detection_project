```python
import pandas as pd
from imblearn.over_sampling import SMOTE
```

```python
X_test = pd.read_csv('../data/X_test.csv')
y_test = pd.read_csv('../data/y_test.csv')
```

```python
smote = SMOTE(sampling_strategy=1, random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_test, y_test)
```

```python
X_test.shape
```

```python
y_test['Class'].value_counts()
```

```python
X_resampled.shape
```

```python
y_resampled['Class'].value_counts()
```

```python
df_resampled = pd.concat([X_resampled, y_resampled], axis=1)
```

```python
df_resampled.rename(columns={'Class': 'is_fraud'}, inplace=True)
```

```python
positive_data = df_resampled[df_resampled['is_fraud'] == 1]
negative_data = df_resampled[df_resampled['is_fraud'] == 0]
```

```python
positive_data
```

```python
positive_data.to_csv('../data/fraud_data.csv', index=False)
negative_data.to_csv('../data/legit_data.csv', index=False)
```
